```python
from __future__ import print_function

import pandas as pd
import datetime
import math
from ipywidgets import interact
from bokeh.io import push_notebook
from bokeh.document import Document
from bokeh.plotting import show,figure,output_notebook
from bokeh.models import Range1d, ColumnDataSource, LinearAxis, PanTool, WheelZoomTool,HoverTool,VBox,HBox,Select,DataRange1d
output_notebook()
```

```python
df_perform = pd.read_csv('npp_performance.csv', index_col='Date')
df_perform['average_perform'] = df_perform.mean(axis=1)
locations = list(df_perform.columns.values)
location = 'Arkansas Nuclear 1'
x=df_perform.index.to_datetime()
y=df_perform[location]
y2=df_perform['average_perform']
```

```python
source = ColumnDataSource(data=dict(x=x,y=y))

plot = figure(x_axis_type="datetime",tools="save,wheel_zoom,pan",plot_width=1000,plot_height=300)
plot.line(x,y,line_color="green",source=source)
plot.line(x,y2,line_color="red")
plot.x_range = DataRange1d(range_padding=0.0, bounds=None)

def performance(location):
    source.data['y'] = df_perform[location]
    plot.title=source.data['y'].name + ' Plant Performance'
    source.push_notebook()
```

```python
show(plot)
```

```python
interact(performance,location=locations)
```
